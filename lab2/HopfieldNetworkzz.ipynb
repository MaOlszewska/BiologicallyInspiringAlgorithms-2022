```python
def generateTest(numbers,  p):
    
    for number in range(len(numbers)):
        t = []
        while len(t) < p:
            x = random.randint(0, 9)
            y = random.randint(0, 9)
            if  not (x, y) in t:
                t.append((x, y))
                z = numbers[number][x][y]
                numbers[number][x][y] = -1 if z == 1 else 1

        plt.matshow(np.reshape(numbers[number], (-1, 10)), cmap = plt.cm.binary)

    
```

```python
generateTest(samplePattern(),10)
```

```python
def calculate_hebb(size):
  neurons_range = range(size**2)
  W = [[0 for _ in neurons_range] for _ in neurons_range]
  n = len(W)
  lp = len(sample[0])
  for xw in range(n):
    for yw in range(n):
      if xw != yw:
        su = 0
        for pat in range(len(sample)):
          su += sample[pat][yw%lp][yw//lp] * sample[pat][xw%lp][xw//lp]
        su /= n
        W[xw][yw] = su
        W[yw][xw] = su
  return W

```

```python
size = 5
w = calculate_hebb(size)
print(w)
```
